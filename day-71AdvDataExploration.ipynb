# Sorting Values & Adding Columns

## Lowest Risk Majors

```python
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

## Insert a column into the dataframe

```python
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

## Sorting values by column

```python
low_risk = clean_df.sort_values('Spread')
```

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

## Challenge II

### Find the degrees with the highest potential earnings.

```python
high_potential_earnings = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
```

```python
high_potential_earnings[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

### Find the degrees with the greatest spread in salaries.
